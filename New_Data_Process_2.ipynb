```python
from glob import glob
import os
import SimpleITK as itk
import matplotlib.pyplot as plt
from scipy.misc import imsave
import cv2

dcm_root_path='C:\\Users\\ilab\\Desktop\\Test\\Dicoms'
img_root_path='C:\\Users\\ilab\\Desktop\\Test\\Images'

patient_index = 1

# ADC, DCE
# series = 'T2'
# series = 'ADC'
# series = 'DCE'

series_list = ['T2','ADC','DCE']

def save_imgs(series,patient_index):

    dcm_folders = os.listdir(dcm_root_path)

    dcm_folder = dcm_folders[patient_index]

    dcm_path = dcm_root_path +'\\'+ dcm_folder+ '\\' +series+ '\\'

    img_name_path =  img_root_path +'\\'+ dcm_folder

    img_series_path = img_root_path + '\\'+ dcm_folder+ '\\Image\\' +series+ '\\'

    print (img_series_path)

    dcms = glob(dcm_path+'*.dcm')



    dcm_name = dcms[0].split('\\')[-3]

    if not os.path.exists(img_name_path):
        os.mkdir(img_name_path)


    
    if not os.path.exists(img_root_path + '\\'+ dcm_folder+ '\\Image\\'):
        os.mkdir(img_root_path + '\\'+ dcm_folder+ '\\Image')
        os.mkdir(img_root_path + '\\'+ dcm_folder+ '\\Mask')
    if not os.path.exists(img_root_path + '\\'+ dcm_folder+ '\\Mask\\'+series):
        os.mkdir(img_root_path + '\\'+ dcm_folder+ '\\Mask\\'+series)

        for i in range(10):
            os.mkdir(img_root_path + '\\'+ dcm_folder+ '\\Mask\\'+series+'\\'+(str)(i+1))
        os.mkdir(img_root_path + '\\'+ dcm_folder+ '\\Mask\\'+series+'\\BPH')

    if not os.path.exists(img_series_path):
        os.mkdir(img_series_path)
        
        
    dcms


    for i in range(len(dcms)):
        itkimage = itk.ReadImage(dcms[i])
        print (dcms[i])
        img_array = itk.GetArrayFromImage(itkimage)[0]  
        if i<9:
            imsave(img_series_path+'slice_00'+ (str)(i+1)+'.jpg',img_array)
            img_array = cv2.resize(cv2.imread(img_series_path+'slice_00'+ (str)(i+1)+'.jpg',-1),(640,640))
            imsave(img_series_path+'slice_00'+(str)(i+1)+'.jpg',img_array)
        else:
            imsave(img_series_path+'slice_0'+ (str)(i+1)+'.jpg',img_array)
            img_array = cv2.resize(cv2.imread(img_series_path+'slice_0'+ (str)(i+1)+'.jpg',-1),(640,640))
            imsave(img_series_path+'slice_0'+(str)(i+1)+'.jpg',img_array)



    print (img_array.shape)

    plt.imshow(img_array,cmap=plt.cm.gray)
    plt.show()


dicoms = os.listdir(dcm_root_path)

for patient_index in range(len(dicoms)):
    try:
        for series in series_list:
            save_imgs(series,patient_index)
    except:
        with open('Log.txt','a+') as f:
            f.write(dicoms[patient_index])
        pass
```
